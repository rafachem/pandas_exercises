### Step 9. Create a function called majority that returns a boolean value to a new column called legal_drinker (Consider majority as older than 17 years old)

```python
def majority(df):
    df['called legal_drinker'] = df['age'] > 17
    return df['called legal_drinker']
```

```python
df_new['called legal_drinker'] = majority(df_new)
```

```python
df_new
```

```python
#Solución profesor
def majority2(x):
    if x > 17:
        return True
    else:
        return False
```

```python
df_new['called legal_drinker'] = df_new['age'].apply(majority2)
df_new
```

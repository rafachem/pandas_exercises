### Step 3. Assign it to a variable called users.

```python
df_users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|')
```

### Step 4. Discover what is the mean age per occupation

```python
df_users.groupby(by='occupation')[['age']].mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
total_gender_occupation = df_users.groupby(by=['occupation'])[['gender']].count()
```

```python
df_users[df_users['gender'] == 'M'].groupby(['occupation'])[['gender']].count()/total_gender_occupation*100
```

```python
#Solución clase:
res = df_users.groupby(['occupation','gender']).size().groupby('occupation').apply(lambda x:x/x.sum())
res
```

```python
type(res.reset_index(name='% male').query('gender == "M"'))
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
df_users.groupby(by='occupation')[['age']].min()+df_users.groupby(by='occupation')[['age']].max()
```

```python
pd.concat([df_users.groupby(by='occupation')[['age']].min(),df_users.groupby(by='occupation')[['age']].max()], axis=1,keys=['min', 'max'])
```

```python
df_users.groupby('occupation').age.agg(['min', 'max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
df_users.groupby(['occupation','gender'])[['age']].mean()
```

### Step 8.  For each occupation present the percentage of women and men

```python
total_age_per_occupation = df_users.groupby(['occupation'])[['gender']].count()
df_users.groupby(['occupation','gender'])[['gender']].count()/total_age_per_occupation*100
```
